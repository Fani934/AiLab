```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Defining the dataset as per Table 1
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Displaying the dataset
print(df)

# Creating a scatter plot of area vs price
%matplotlib inline
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.scatter(df.area, df.price, color='red', marker='+')

# Preparing the data for linear regression
X = df.drop('price', axis='columns')  # Feature (independent variable)
y = df.price  # Target (dependent variable)

# Creating and training the linear regression model
reg = LinearRegression()
reg.fit(X, y)

# Predicting the price for homes with different areas
areas_to_predict = [[5000], [8000], [9000]]
predicted_prices = reg.predict(areas_to_predict)

# Printing the predictions
for area, price in zip(areas_to_predict, predicted_prices):
    print(f"Predicted price for {area[0]} sq ft: ${price:.2f}")

# Coefficient and intercept
print(f"Coefficient (m): {reg.coef_[0]}")
print(f"Intercept (b): {reg.intercept_}")

# Verifying the equation Y = mX + b for one example
manual_prediction = 5000 * reg.coef_[0] + reg.intercept_
print(f"Manual Prediction for 5000 sq ft: ${manual_prediction:.2f}")

# Plotting the regression line with scatter points
plt.plot(df.area, reg.predict(X), color='blue', label='Regression Line')
plt.legend()
plt.show()
```
