```python
# Importing the required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Reading the dataset
df = pd.read_csv('SaloonShop.csv')

# Cleaning column names (removing any extra spaces)
df.columns = df.columns.str.strip()

# Displaying the cleaned dataset
print("Dataset:")
print(df)

# Creating a scatter plot of Customers vs Daily Revenue
%matplotlib inline
plt.xlabel('Customers')
plt.ylabel('Daily Revenue ($)')
plt.scatter(df['Customers'], df['Daily Revenue ($)'], color='red', marker='+', label='Actual Data')

# Preparing the data for linear regression
X = df[['Customers']]  # Feature (independent variable)
y = df['Daily Revenue ($)']  # Target (dependent variable)

# Creating and training the linear regression model
reg = LinearRegression()
reg.fit(X, y)

# Plotting the regression line with scatter points
plt.plot(df['Customers'], reg.predict(X), color='blue', label='Regression Line')
plt.legend()
plt.show()

# Predicting daily revenue for specific numbers of customers
customers_to_predict = [[20], [40], [60]]  # Example: Predict for 20, 40, and 60 customers
predicted_revenues = reg.predict(customers_to_predict)

# Printing the predictions
print("\nPredicted Daily Revenues:")
for customers, revenue in zip(customers_to_predict, predicted_revenues):
    print(f"Predicted daily revenue for {customers[0]} customers: ${revenue:.2f}")

# Coefficient and intercept
print("\nModel Parameters:")
print(f"Coefficient (m): {reg.coef_[0]}")
print(f"Intercept (b): {reg.intercept_}")

# Verifying the linear regression equation Y = mX + b for one example
example_customers = 20
manual_prediction = example_customers * reg.coef_[0] + reg.intercept_
print(f"\nManual Calculation for {example_customers} customers: ${manual_prediction:.2f}")
```
